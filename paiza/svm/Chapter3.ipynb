```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
# 再生した音声の波形を描画する
a, sr = librosa.load('voiceset/kirishima_b01.wav')
librosa.display.waveplot(a, sr)
```

横軸が時間、縦軸が音の大きさを表す振幅

```python
# 波形を表すデータ：振幅の数値の列
print(a)
print(len(a))

# サンプリングレート -> 1秒間のフレーム数
print(sr)
```

```python
# 6秒の音声なので、次がlen(a)と一致するはず
print(sr * 6)
```

```python
# 高音, 低音
files = [['sample/hi.wav', '高音'], ['sample/lo.wav', '低音']]
for f in files:
    a, sr = librosa.load(f[0])
    librosa.display.waveplot(a, sr)
    print(f[1])
    plt.show()
```

```python
# 高音, 低音
files = [['sample/hi.wav', '高音'], ['sample/lo.wav', '低音']]
for f in files:
    a, sr = librosa.load(f[0])
    librosa.display.waveplot(a[0:100], sr)
    print(f[1])
    plt.show()

print('音声データは波形で表現できる')
print('時間あたりの振動数が多いと高くなり、少ないと低くなる')
```
